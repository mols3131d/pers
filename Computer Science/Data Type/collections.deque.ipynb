# [collections.deque](https://docs.python.org/ko/3/library/collections.html#deque-objects)

```python
from collections import deque

dq = [1, 2, 3, 4, 5]
dq = deque(dq)
print(dq)
dq
```

```python
dir(dq)
```

## 요소 찾기
- `index(x[, start[, stop]])`

    데크에 있는 x의 위치를 반환합니다.<br>
    (인덱스 start 또는 그 이후, 그리고 인덱스 stop 이전). 

```python
dq.index(1)
```

```python
dq.index(3, 0, 4)
```

## 요소 넣기
- `append(x)`

    오른쪽에 `x`를 추가합니다.

- `appendleft(x)`
    
    왼쪽에 `x`를 추가합니다.

- `insert(i, x)`

    x를 데크의 i 위치에 삽입합니다.

- `extend(iterable)`

    iterable 인자에서 온 요소를 추가하여 데크의 오른쪽을 확장합니다.

- `extendleft(iterable)`

    iterable에서 온 요소를 추가하여 데크의 왼쪽을 확장합니다. 일련의 왼쪽 추가는 iterable 인자에 있는 요소의 순서를 뒤집는 결과를 줍니다.

```python
dq.append(6)
dq
```

```python
dq.appendleft(0)
dq
```

```python
dq.insert(2, 100)
dq
```

## 요소 빼기
- `pop()`
    오른쪽에서 요소를 제거하고 반환.
- `popleft()`
    왼쪽에서 요소를 제거하고 반환.
- `remove(value)`
    왼쪽 부터 첫 번째의 `value`를 제거.

```python
dq.pop()
```

```python
dq
```

```python
dq.popleft()
```

```python
dq
```

```python
dq.remove(100)
dq
```
