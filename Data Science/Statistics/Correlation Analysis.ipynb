```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets

data = datasets.load_diabetes()
df = pd.DataFrame(data["data"], index=data["target"], columns=data["feature_names"])
df
```

```python
X = df.bmi.values
Y = df.index.values

plt.scatter(X, Y, alpha=0.5)
plt.title("TARGET ~ BMI")
plt.xlabel("BMI")
plt.ylabel("TARGET")
plt.show()
```

```python
cov = (np.sum(X * Y) - len(X) * np.mean(X) * np.mean(Y)) / len(X)
cov
```

```python
cov = np.cov(X, Y)
cov
```

```python
corr = cov / (np.std(X) * np.std(Y))
corr
```

```python
corr = np.corrcoef(X,Y)
corr
```

```python
from scipy.stats import pearsonr

pval = pearsonr(X, Y)[1]
pval
```

```python
with pd.option_context("display.float_format", '{:.4f}'.format):
    display(df.corr())
```

```python
mask = np.zeros_like(df.corr(), dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(14, 14))
sns.heatmap(data=df.corr(), annot=True, fmt=".4f", mask=mask)
```

```python
from scipy.stats import pearsonr

# pd.options.display.float_format = '{:.5f}'.format

with pd.option_context("display.float_format", '{:.4f}'.format):
    pval = df.corr(method=lambda x, y: pearsonr(x, y)[1])
    display(pval)
```

```python
from scipy.stats import pearsonr

# pd.options.display.float_format = '{:.5f}'.format

with pd.option_context("display.float_format", '{:.4f}'.format):
    pval = df.corr(method=lambda x, y: pearsonr(x, y)[1])
    display(pval)
    
mask = np.zeros_like(df.corr(), dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(14, 14))
sns.heatmap(data=pval, annot=True, fmt=".4f")
```

```python
l1 = ['age','sex','bmi','bp']
col = df.columns.to_list()

for i in l1:
    fig, axes = plt.subplots(len(col)//len(l1)+1, len(l1), figsize=(20, 10), constrained_layout=True)
    for a1, j in enumerate(col):
        sns.regplot(data=df, x=i, y=j, ax=axes[a1 // len(l1), a1 % len(l1)])
        axes[a1 // 4, a1 % 4].set_title(f"x={i}, y={j}", fontsize=12)

    fig.suptitle(f"regplot, x={i}", fontsize=20)
    plt.show()
```
