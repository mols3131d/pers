```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 데이터를 생성합니다.
data = np.array([[1, 2, 3], [3, 4, 9], [5, 4, 12], [7, 6, 15]])

# 데이터를 표준화합니다.
scaler = StandardScaler()
data_std = scaler.fit_transform(data)

# 주성분 분석에 사용할 차원 수를 정의합니다.
n_components = 2

# PCA를 생성하고 훈련 데이터에 적용합니다.
pca = PCA(n_components=n_components)
data_pca = pca.fit_transform(data_std)

print("설명된 분산 :", pca.explained_variance_ratio_)
print("주성분 :", pca.components_)
print("투영된 데이터(주성분 점수) :", data_pca)

# 시각화를 위한 코드입니다.
fig, ax = plt.subplots()

# 각 점을 그래프에 표시합니다.
ax.scatter(data_pca[:, 0], data_pca[:, 1], label="Projected Data")

# 주성분 축의 번호와 함께 좌표를 표시합니다.
for i, txt in enumerate(data_pca):
    ax.annotate(f"{i + 1}", (data_pca[i, 0], data_pca[i, 1]), fontsize=12)

# 그래프의 축 이름을 설정합니다.
ax.set_xlabel("First Principal Component")
ax.set_ylabel("Second Principal Component")

plt.legend()
plt.show()
```
