```python
from collections import defaultdict

d = defaultdict(list)

print(d)

for i in range(8):
    d[i] += ["a"]
    
print(d)
```

## `collections.Counter`

```python
from collections import Counter

#   
sample_list = ['a', 'b', 'c', 'a', 'b', 'a']
counter_list = Counter(sample_list)
print(counter_list)  # : Counter({'a': 3, 'b': 2, 'c': 1})

#    
sample_str = 'abracadabra'
counter_str = Counter(sample_str)
print(counter_str)  # : Counter({'a': 5, 'b': 2, 'r': 2, 'c': 1, 'd': 1})

#     
most_common = counter_str.most_common(2)
print(most_common)  # : [('a', 5), ('b', 2)]

#   
counter_diff = counter_list - counter_str
print(counter_diff)  # : Counter({'c': -1})
```

```python
from collections import Counter

counter1 = Counter(a=3, b=2, c=1)
counter2 = Counter(a=1, b=2, c=3)

#  
intersection = counter1 & counter2
print(intersection)  # : Counter({'a': 1, 'b': 2, 'c': 1})

#  
union = counter1 | counter2
print(union)  # : Counter({'a': 3, 'b': 2, 'c': 3})
```

## [`collections.deque`](https://docs.python.org/3.8/library/collections.html#collections.deque)

### deque(iterable, maxlen) - 
- Parameters
  - iterable
    - accept argument dtype
      - string
      - tuple
      - list
      - set
      - dictionary (key)
  - maxlen
    - deque     .

```python
from collections import deque

dq = deque(iterable=[1,2,3,4,5], maxlen=None)
print(dq)
```

```python
dir(dq)
```

### dq.copy() - 

```python
dq = deque([1,2,3,4,5])
dq2 = dq.copy()
print(dq)
print(dq2)
```

### dq.index(elem, start, stop)
- Params
  - elem
    -  
  - start
    -   
  - stop
    -   
- Error
  - ValueError
    -    elem   

```python
dq = deque([3,1,2,3,4,5,3])
print(dq)

elem = 3
print(dq.index(elem))
print(dq.index(elem, 4))
```

### dq.reverse() - dq   .

```python
dq = deque([1,2,3,4,5])
print(dq)

dq.reverse()
print(dq)
```

### dq.pop() -     

```python
dq = deque([1,2,3,4,5])
print(dq)

dq.pop()
print(dq)

print(dq.pop())
print(dq)

pop = dq.pop()
print(dq)

print(pop)
print(dq)
```
