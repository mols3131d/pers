# `collections.defaultdict()`

```python
# 기본 내장 딕셔너리

d = {}

print(d)

for i in range(8):
    try:
        d[i] += ["a"]
    except:
        print("err")
    
print(d)
```

```python
from collections import defaultdict

d = defaultdict(list)

print(d)

for i in range(8):
    d[i] += ["a"]
    
print(d)
```
